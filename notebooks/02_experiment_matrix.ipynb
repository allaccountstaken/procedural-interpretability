# 02 — Experiment Matrix

Runs the eight-condition matrix across sample tasks and stores traces in `results/`.

```
from dotenv import load_dotenv
load_dotenv()

from procedural_interp.runner import run_experiment

# This will call the OpenAI API once per task-condition combination.
# For 4 sample tasks x 8 conditions = 32 model calls.
# Uncomment when ready.
# traces = run_experiment("../configs/experiment_openai_demo.yaml")
```

```
import json
from pathlib import Path
import pandas as pd

path = Path("../results/openai_demo_001.jsonl")
if path.exists():
    rows = [json.loads(line) for line in path.read_text().splitlines() if line.strip()]
    df = pd.DataFrame(rows)
    display(df[["task_id", "model_condition", "context_condition", "checklist_enabled", "final_action", "misuse_result", "misalignment_result", "mistake_result"]].head())
else:
    print("No results yet. Run the experiment cell above first.")
```

```
if path.exists():
    summary = (
        df.groupby(["model_condition", "context_condition", "checklist_enabled", "final_action"])
          .size()
          .reset_index(name="n")
          .sort_values(["model_condition", "context_condition", "checklist_enabled", "final_action"])
    )
    display(summary)
```

```
if path.exists():
    import matplotlib.pyplot as plt
    counts = df.groupby(["checklist_enabled", "final_action"]).size().unstack(fill_value=0)
    counts.plot(kind="bar")
    plt.title("Final actions by checklist condition")
    plt.xlabel("Checklist enabled")
    plt.ylabel("Count")
    plt.tight_layout()
```
